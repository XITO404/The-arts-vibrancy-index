```python
import pandas as pd

# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}
# 데이터 프레임 생성
Busan_Arts_Index = pd.DataFrame(data)
Busan_Arts_Index
```

```python
busan_arts_creation_production_index = pd.read_csv('busan_arts_creation_production_index.csv')
busan_arts_finance_index = pd.read_csv('busan_arts_finance_index.csv')
busan_arts_facilities_index = pd.read_csv('busan_arts_facilities_index.csv')
busan_arts_enjoyment_index = pd.read_csv('busan_arts_enjoyment_index.csv')
busan_arts_achievement_index = pd.read_csv('busan_arts_achievement_index.csv')
```

## 예술창작/생산지수

```python
Busan_Arts_Index['예술창작/생산지수'] = busan_arts_creation_production_index['busan_arts_creation_production_index'].round(2)
Busan_Arts_Index
```

## 예술재정지수

```python
Busan_Arts_Index['예술재정지수'] = busan_arts_finance_index['busan_arts_finance_index'].round(2)
Busan_Arts_Index
```

## 예술시설지수

```python
Busan_Arts_Index['예술시설지수'] = busan_arts_facilities_index['busan_arts_facilities_index'].round(2)
Busan_Arts_Index
```

## 예술향유지수

```python
Busan_Arts_Index['예술향유지수'] = busan_arts_enjoyment_index['busan_arts_enjoyment_index'].round(2)
Busan_Arts_Index
```

## 예술성과지수

```python
Busan_Arts_Index['예술성과지수'] = busan_arts_achievement_index['busan_arts_achievement_index'].round(2)
Busan_Arts_Index
```

## Busan arts index

```python
Busan_Arts_Index['부산예술활력지수'] = busan_arts_creation_production_index['busan_arts_creation_production_index'] + busan_arts_finance_index['busan_arts_finance_index'] + busan_arts_facilities_index['busan_arts_facilities_index'] +  busan_arts_enjoyment_index['busan_arts_enjoyment_index'] + busan_arts_achievement_index['busan_arts_achievement_index']
Busan_Arts_Index['부산예술활력지수'] = (Busan_Arts_Index['부산예술활력지수']/5).round(2)
Busan_Arts_Index
```

```python
summary = Busan_Arts_Index.describe()
summary
```

```python
import matplotlib.pyplot as plt

# 폰트 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'

Busan_Arts_Index['부산예술활력지수'].plot(kind='bar')
plt.title('부산예술활력지수')

# x축에 구별 이름 추가
districts = ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구']
plt.xticks(range(len(districts)), districts, rotation=45, ha='right')

plt.show()
```

```python
Busan_Arts_Index = Busan_Arts_Index.sort_values(by='부산예술활력지수', ascending=False)
Busan_Arts_Index
```
