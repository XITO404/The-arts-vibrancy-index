## 문화예술만족도 및 관람의향 (Arts and culture satisfaction and intent to attend)

```python
# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
Average_satisfaction_mean = arts_achievement_index['avg_satisfaction'].mean()
Average_satisfaction_mean
```

```python
cal_df['avg_satisfaction'] = arts_achievement_index['avg_satisfaction']/Average_satisfaction_mean
cal_df
```

```python
Sum_of_willingness_to_watch_mean = arts_achievement_index['intent_to_watch'].mean()
Sum_of_willingness_to_watch_mean
```

```python
cal_df['intent_to_watch'] = arts_achievement_index['intent_to_watch']/Sum_of_willingness_to_watch_mean
cal_df
```

```python
cal_df['sat_intent'] = (cal_df['avg_satisfaction'] + cal_df['intent_to_watch'])/2
cal_df['sat_intent'] = cal_df['sat_intent']*100
cal_df
```

```python
final_art_achievement_index['sat_intent'] = cal_df['sat_intent']
final_art_achievement_index
```

```python
final_art_achievement_index['sat_intent'].describe()
```

## 문화예술교육만족도 및 수강의향

```python
Satisfaction_intention_arts_education_mean = arts_achievement_index['edu_intent'].mean()
Satisfaction_intention_arts_education_mean
```

```python
final_art_achievement_index['edu_intent'] = (arts_achievement_index['edu_intent']/Satisfaction_intention_arts_education_mean)*100
final_art_achievement_index
```

## 여가생활 만족도

```python
Leisure_satisfaction_mean = arts_achievement_index['sat_leisure'].mean()
Leisure_satisfaction_mean
```

```python
final_art_achievement_index['sat_leisure'] = (arts_achievement_index['sat_leisure']/Leisure_satisfaction_mean)*100
final_art_achievement_index['sat_leisure'] = final_art_achievement_index['sat_leisure'].round(2)
final_art_achievement_index
```

## 행복 및 삶의질 (Happiness and quality of life)

```python
cal_df['Happiness_level'] = arts_achievement_index['happiness']
cal_df['Quality_of_life'] = arts_achievement_index['life']
cal_df
```

```python
Happiness_level_mean= arts_achievement_index['happiness'].mean()
Happiness_level_mean
```

```python
Quality_of_life_mean = arts_achievement_index['life'].mean()
Quality_of_life_mean
```

```python
cal_df['Happiness_level'] = (arts_achievement_index['happiness']/Happiness_level_mean)*100
cal_df['Quality_of_life'] = (arts_achievement_index['life']/Quality_of_life_mean)*100
cal_df
```

```python
final_art_achievement_index['happy_life'] = (cal_df['Happiness_level'] + cal_df['Quality_of_life'])/2
final_art_achievement_index['happy_life'] = final_art_achievement_index['happy_life'].round(2)
final_art_achievement_index
```

## 공연시장매출액

```python
arts_achievement_index
```

```python
Performance_market_revenue_mean = arts_achievement_index['perf_revenue'].mean()
Performance_market_revenue_mean
```

```python
final_art_achievement_index['perf_revenue'] = (arts_achievement_index['perf_revenue']/Performance_market_revenue_mean)*100
final_art_achievement_index
```

## 미술시장매출액

```python
Art_market_revenue_mean = arts_achievement_index['art_revenue'].mean()
Art_market_revenue_mean
```

```python
final_art_achievement_index['art_revenue'] = (arts_achievement_index['art_revenue']/Art_market_revenue_mean)*100
final_art_achievement_index
```

## 문화예술 국제교류 건수

```python
Cultural_exchanges_mean = arts_achievement_index['global_exchange'].mean()
Cultural_exchanges_mean
```

```python
final_art_achievement_index['global_exchange'] = (arts_achievement_index['global_exchange']/Cultural_exchanges_mean)*100
final_art_achievement_index
```

## 향유결과지수

```python
final_art_achievement_index['enjoyment_outcome_index'] = (final_art_achievement_index['sat_intent'] + final_art_achievement_index['edu_intent'] + final_art_achievement_index['sat_leisure'] + final_art_achievement_index['happy_life'])/4
final_art_achievement_index['enjoyment_outcome_index'] = final_art_achievement_index['enjoyment_outcome_index'].round(2)
final_art_achievement_index
```

## 시장성과지수

```python
final_art_achievement_index['market_performance_index'] = (final_art_achievement_index['perf_revenue'] + final_art_achievement_index['art_revenue'] + final_art_achievement_index['global_exchange'])/3
final_art_achievement_index
```

## 예술성과지수

```python
final_art_achievement_index['busan_arts_achievement_index'] = (final_art_achievement_index['enjoyment_outcome_index'] + final_art_achievement_index['market_performance_index'])/2
final_art_achievement_index['busan_arts_achievement_index'] = final_art_achievement_index['busan_arts_achievement_index'].round(2)
final_art_achievement_index
```
