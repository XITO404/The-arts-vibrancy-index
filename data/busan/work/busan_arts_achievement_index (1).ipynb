```python
import pandas as pd
import numpy as np
```

```python
# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}
# 데이터 프레임 생성
final_art_achievement_index = pd.DataFrame(data)
final_art_achievement_index
```

```python
arts_achievement_index = pd.read_csv('../Busan_data/arts_achievement_index_calculate.csv')
arts_achievement_index
```

## 문화예술만족도 및 관람의향 (Arts and culture satisfaction and intent to attend)

```python
# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
Average_satisfaction_mean = arts_achievement_index['Average_satisfaction'].mean()
Average_satisfaction_mean
```

```python
cal_df['Average_satisfaction'] = arts_achievement_index['Average_satisfaction']/Average_satisfaction_mean
cal_df
```

```python
Sum_of_willingness_to_watch_mean = arts_achievement_index['Sum_of_willingness_to_watch'].mean()
Sum_of_willingness_to_watch_mean
```

```python
cal_df['Sum_of_willingness_to_watch'] = arts_achievement_index['Sum_of_willingness_to_watch']/Sum_of_willingness_to_watch_mean
cal_df
```

```python
cal_df['Arts_and_culture_satisfaction_and_intent_to_attend'] = (cal_df['Average_satisfaction'] + cal_df['Sum_of_willingness_to_watch'])/2
cal_df['Arts_and_culture_satisfaction_and_intent_to_attend'] = cal_df['Arts_and_culture_satisfaction_and_intent_to_attend']*100
cal_df
```

```python
final_art_achievement_index['Arts_and_culture_satisfaction_and_intent_to_attend'] = cal_df['Arts_and_culture_satisfaction_and_intent_to_attend']
final_art_achievement_index
```

## 문화예술교육만족도 및 수강의향 

```python
Satisfaction_intention_arts_education_mean = arts_achievement_index['Satisfaction_intention_arts_education'].mean()
Satisfaction_intention_arts_education_mean
```

```python
final_art_achievement_index['Satisfaction_and_intention_to_take_arts_education'] = (arts_achievement_index['Satisfaction_intention_arts_education']/Satisfaction_intention_arts_education_mean)*100
final_art_achievement_index
```

## 여가생활 만족도

```python
Leisure_satisfaction_mean = arts_achievement_index['Leisure_satisfaction'].mean()
Leisure_satisfaction_mean
```

```python
final_art_achievement_index['Satisfaction_with_leisure_time'] = (arts_achievement_index['Leisure_satisfaction']/Leisure_satisfaction_mean)*100
final_art_achievement_index
```

## 행복 및 삶의질 (Happiness and quality of life)

```python
# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
cal_df['Happiness_level'] = arts_achievement_index['Happiness_level']
cal_df['Quality_of_life'] = arts_achievement_index['Quality_of_life']
cal_df
```

```python
Happiness_level_mean= arts_achievement_index['Happiness_level'].mean()
Happiness_level_mean
```

```python
Quality_of_life_mean = arts_achievement_index['Quality_of_life'].mean()
Quality_of_life_mean
```

```python
cal_df['Happiness_level'] = (arts_achievement_index['Happiness_level']/Happiness_level_mean)*100
cal_df['Quality_of_life'] = (arts_achievement_index['Quality_of_life']/Quality_of_life_mean)*100
cal_df
```

```python
final_art_achievement_index['Happiness_and_quality_of_life'] = (cal_df['Happiness_level'] + cal_df['Quality_of_life'])/2
final_art_achievement_index
```

## 공연시장매출액

```python
arts_achievement_index
```

```python
Performance_market_revenue_mean = arts_achievement_index['Performance_market_revenue'].mean()
Performance_market_revenue_mean
```

```python
final_art_achievement_index['Performance_marketplace_sales'] = (arts_achievement_index['Performance_market_revenue']/Performance_market_revenue_mean)*100
final_art_achievement_index
```

## 미술시장매출액

```python
Art_market_revenue_mean = arts_achievement_index['Art_market_revenue'].mean()
Art_market_revenue_mean
```

```python
final_art_achievement_index['Art_market_sales'] = (arts_achievement_index['Art_market_revenue']/Art_market_revenue_mean)*100
final_art_achievement_index
```

## 문화예술 국제교류 건수

```python
Cultural_exchanges_mean = arts_achievement_index['Cultural_exchanges'].mean()
Cultural_exchanges_mean
```

```python
final_art_achievement_index['Number_of_international_cultural_exchanges'] = (arts_achievement_index['Cultural_exchanges']/Cultural_exchanges_mean)*100
final_art_achievement_index
```

## 향유결과지수

```python
final_art_achievement_index['Enjoyment_Outcome_Index'] = (final_art_achievement_index['Arts_and_culture_satisfaction_and_intent_to_attend'] + final_art_achievement_index['Satisfaction_and_intention_to_take_arts_education'] + final_art_achievement_index['Satisfaction_with_leisure_time'] + final_art_achievement_index['Happiness_and_quality_of_life'])/4
final_art_achievement_index
```

## 시장성과지수

```python
final_art_achievement_index['Market_Performance_Index'] = (final_art_achievement_index['Performance_marketplace_sales'] + final_art_achievement_index['Art_market_sales'] + final_art_achievement_index['Number_of_international_cultural_exchanges'])/3
final_art_achievement_index
```

## 예술성과지수

```python
final_art_achievement_index['art_achievement_index'] = (final_art_achievement_index['Enjoyment_Outcome_Index'] + final_art_achievement_index['Market_Performance_Index'])/2
final_art_achievement_index
```

```python
final_art_achievement_index.to_csv('busan_arts_achievement_index.csv', encoding='utf-8-sig', index=False)
```
