```python
import pandas as pd
import numpy as np
```

```python
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
final_arts_finance_index = pd.DataFrame(data)
final_arts_finance_index
```

```python
art_finance_index = pd.read_csv('../Seoul_data/arts_finance_index_calculate.csv')
art_finance_index
```

## 문화부 예산

```python
final_arts_finance_index['CulBudget'] = art_finance_index['CulBudget_2022'] / art_finance_index['CulBudget_2021']
final_arts_finance_index['CulBudget'] = final_arts_finance_index['CulBudget']*100
final_arts_finance_index
```

## 정부 예산 대비 문화부 예산

```python
final_arts_finance_index['GbvsCb'] = art_finance_index['GbvsCb_2022'] / art_finance_index['GbvsCb_2021']
final_arts_finance_index['GbvsCb'] = final_arts_finance_index['GbvsCb']*100
final_arts_finance_index
```

## 문화예술 예산

```python
final_arts_finance_index['CulArtBudget'] = art_finance_index['CulArtBudget_2022'] / art_finance_index['CulArtBudget_2021']
final_arts_finance_index['CulArtBudget'] = final_arts_finance_index['CulArtBudget']*100
final_arts_finance_index
```

## 문화부 예산 대비 문화예술 예산

```python
final_arts_finance_index['MofCBvsAabdCB'] = art_finance_index['MofCBvsAabdCB_2022'] / art_finance_index['MofCBvsAabdCB_2021']
final_arts_finance_index['MofCBvsAabdCB'] = final_arts_finance_index['MofCBvsAabdCB']*100
final_arts_finance_index
```

## 기업의 문화예술 기부액 및 문화예술 기부액 비중

```python
art_finance_index['CompanyDonation_2021']
```

```python
art_finance_index['CompanyDonation_2022']
```

```python
art_finance_index['CompanyDonation_2022'] / art_finance_index['CompanyDonation_2021']
```

```python
final_arts_finance_index['CompanyDonation'] = art_finance_index['CompanyDonation_2022'] / art_finance_index['CompanyDonation_2021']
final_arts_finance_index['CompanyDonation'] = final_arts_finance_index['CompanyDonation']*100
final_arts_finance_index
```

### Nan -> 0으로 변경

```python
final_arts_finance_index['CompanyDonation'] = final_arts_finance_index['CompanyDonation'].fillna(0)
```

```python
final_arts_finance_index
```

## 예산지수

```python
final_arts_finance_index['Budget_Index'] = final_arts_finance_index['CulBudget'] + final_arts_finance_index['GbvsCb'] + final_arts_finance_index['CulArtBudget'] + final_arts_finance_index['MofCBvsAabdCB']
final_arts_finance_index['Budget_Index'] = final_arts_finance_index['Budget_Index']/4
final_arts_finance_index
```

## 기부지수

```python
final_arts_finance_index['Donation_Index'] = final_arts_finance_index['CompanyDonation']
final_arts_finance_index
```

## 예술재정지수

```python
final_arts_finance_index['Arts_Finance_Index'] = final_arts_finance_index['Budget_Index'] + final_arts_finance_index['Donation_Index']
final_arts_finance_index['Arts_Finance_Index'] = final_arts_finance_index['Arts_Finance_Index']/2
final_arts_finance_index
```

```python
final_arts_finance_index.to_csv('arts_finance_index.csv', encoding='utf-8-sig', index=False)
```
