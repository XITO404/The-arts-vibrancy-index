```python
# 예술 지수 구성 지표 중 예술 향유 지수 table 생성
# 예술 향유지수 = 평균(향유지수, 참여지수)
# 데이터는 소수점 셋째 자리에서 반올림함

import pandas as pd
```

```python
# 계산을 위한 데이터 불러오기
# 천 단위에 들어갔던 콤마 제거
data = pd.read_csv('/content/arts_enjoyment_index_calculate.csv', thousands = ',')
df = pd.DataFrame(data)
districts = df.iloc[:, 0]
districts
```

```python
# 전체 예술 향유지수 데이터 프레임 생성
Arts_enjoyment_index = pd.DataFrame(districts)
Arts_enjoyment_index
```

# 문화예술 관람률

```python
# Audience rate = 100*(audience_rate_2022 / audience_rate_2020)
df['Audience_rate'] = 100*(df['audience_rate_2022']/df['audience_rate_2020'])
df['Audience_rate'] = df['Audience_rate'].round(2)
df
```

```python
Arts_enjoyment_index['audience_rate'] = df['Audience_rate']
Arts_enjoyment_index
```

# 문화예술 관람 횟수

```python
# Visit num = 100*((visits_2022)/(visits_2021))
df['Visit_num'] = 100*(df['visits_2022']/df['visits_2021'])
df['Visit_num'] = df['Visit_num'].round(2)
Arts_enjoyment_index['visit_num'] = df['Visit_num']
Arts_enjoyment_index
```

# 1인당 문화서비스 지출액 및 비중

```python
# Expenditure = 100*[avg{((expenditure_2022)/(expenditure_2021)),((ex_rate_2022)/(ex_rate_2021))}]
df['ex01'] = (df['expenditure_2022']/df['expenditure_2021'])
df['ex02'] = (df['ex_rate_2022']/df['ex_rate_2021'])
df['Expenditure'] = (df['ex01']+df['ex02'])/2*100
df['Expenditure'] = df['Expenditure'].round(2)
df
```

```python
Arts_enjoyment_index['expenditure'] = df['Expenditure']
Arts_enjoyment_index
```

# 독서율 및 도서관 이용률

```python
# Reading = 100*[avg{((num_books_2019)/(num_books_2019)),((library_2019)/(library_2019))}]
# 2021년, 2022년 데이터가 존재하지 않으므로 2019년 데이터 사용
# 그러나 다른 데이터들과 연도가 다르므로 분모와 분자 모두 2019 데이터를 넣어 지표를 100으로 만듦
df['ex03'] = (df['num_books_2019']/df['num_books_2019'])    #1
df['ex04'] = (df['library_2019']/df['library_2019'])        #1
df['Reading'] = (df['ex03']+df['ex04'])/2*100
df
```

```python
Arts_enjoyment_index['reading'] = df['Reading']
Arts_enjoyment_index
```

# 문화예술 자원봉사 경험률

```python
# Volunteer = 100*((volunteer_2022)/(volunteer_2021))
df['Volunteer'] = 100*(df['volunteer_2022']/df['volunteer_2021'])
df['Volunteer'] = df['Volunteer'].round(2)
Arts_enjoyment_index['volunteer'] = df['Volunteer']
Arts_enjoyment_index
```

# 문화예술 동호회 참여율

```python
# Club = 100*((club_2022)/(club_2021))
df['Club'] = 100*(df['club_2022']/df['club_2021'])
df['Club'] = df['Club'].round(2)
Arts_enjoyment_index['club'] = df['Club']
Arts_enjoyment_index
```

# 문화예술교육 경험률

```python
# Education = 100*((education_2021)/(education_2021))
# 2022년 데이터가 존재하지 않으므로 2021년 단독 사용
# 비교가 불가능하므로 분모와 분자 모두 2021 데이터를 넣어 지표를 100으로 만듦

df['Education'] = 100*(df['education_2021']/df['education_2021'])
df
```

```python
Arts_enjoyment_index['education'] = df['Education']
Arts_enjoyment_index
```

# 세부지수 1. 향유지수

```python
# avg(Audience_rate, Visit_num, Expenditure, Reading)
df['Enjoyment_index'] = Arts_enjoyment_index['audience_rate']+Arts_enjoyment_index['visit_num']+Arts_enjoyment_index['expenditure']+Arts_enjoyment_index['reading']
df['Enjoyment_index'] = df['Enjoyment_index']/4
df['Enjoyment_index'] = df['Enjoyment_index'].round(2)
df
```

```python
Arts_enjoyment_index['enjoyment_index'] = df['Enjoyment_index']
Arts_enjoyment_index
```

# 세부지수 2. 참여지수

```python
# avg(Volunteer, Club, Education)
df['Participation_index'] = Arts_enjoyment_index['volunteer']+Arts_enjoyment_index['club']+Arts_enjoyment_index['education']
df['Participation_index'] = df['Participation_index']/3
df['Participation_index'] = df['Participation_index'].round(2)
df
```

```python
Arts_enjoyment_index['participation_index'] = df['Participation_index']
Arts_enjoyment_index
```

# 최종 분야 지수 산출 - 예술향유지수
: Arts enjoyment index

```python
# avg(Enjoyment_index, Participation_index)
Arts_enjoyment_index['arts_enjoyment_index'] = Arts_enjoyment_index['enjoyment_index']+Arts_enjoyment_index['participation_index']
Arts_enjoyment_index['arts_enjoyment_index'] = Arts_enjoyment_index['arts_enjoyment_index']/2
Arts_enjoyment_index['arts_enjoyment_index'] = Arts_enjoyment_index['arts_enjoyment_index'].round(2)
Arts_enjoyment_index['arts_enjoyment_index']
```
