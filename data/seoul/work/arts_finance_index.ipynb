```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
final_arts_finance_index = pd.DataFrame(data)
final_arts_finance_index
```

```python
art_finance_index = pd.read_csv('/content/arts_finance_index_calculate.csv')
art_finance_index
```

## 문화부 예산

```python
final_arts_finance_index['cul_budget'] = art_finance_index['cul_budget_2022'] / art_finance_index['cul_budget_2021']
final_arts_finance_index['cul_budget'] = final_arts_finance_index['cul_budget']*100
final_arts_finance_index
```

## 정부 예산 대비 문화부 예산

```python
final_arts_finance_index['cul_government'] = art_finance_index['cul_government_2022'] / art_finance_index['cul_government_2021']
final_arts_finance_index['cul_government'] = final_arts_finance_index['cul_government']*100
final_arts_finance_index
```

## 문화예술 예산

```python
final_arts_finance_index['cna_budget'] = art_finance_index['cna_budget_2022'] / art_finance_index['cna_budget_2021']
final_arts_finance_index['cna_budget'] = final_arts_finance_index['cna_budget']*100
final_arts_finance_index
```

## 문화부 예산 대비 문화예술 예산

```python
final_arts_finance_index['cna_budget_2'] = art_finance_index['cna_budget_2_2022'] / art_finance_index['cna_budget_2_2021']
final_arts_finance_index['cna_budget_2'] = final_arts_finance_index['cna_budget_2']*100
final_arts_finance_index
```

## 기업의 문화예술 기부액 및 문화예술 기부액 비중

```python
art_finance_index['donation_2021']
```

```python
art_finance_index['donation_2022']
```

```python
art_finance_index['donation_2022'] / art_finance_index['donation_2021']
```

```python
final_arts_finance_index['donation'] = art_finance_index['donation_2022'] / art_finance_index['donation_2021']
final_arts_finance_index['donation'] = final_arts_finance_index['donation']*100
final_arts_finance_index
```

### Nan -> 0으로 변경

```python
final_arts_finance_index['donation'] = final_arts_finance_index['donation'].fillna(0)
```

```python
final_arts_finance_index
```

## 예산지수

```python
final_arts_finance_index['budget_index'] = final_arts_finance_index['cul_budget'] + final_arts_finance_index['cul_government'] + final_arts_finance_index['cna_budget'] + final_arts_finance_index['cna_budget_2']
final_arts_finance_index['budget_index'] = final_arts_finance_index['budget_index']/4
final_arts_finance_index['budget_index'] = final_arts_finance_index['budget_index'].round(2)
final_arts_finance_index
```

## 기부지수

```python
final_arts_finance_index['donation_index'] = final_arts_finance_index['donation']
final_arts_finance_index['donation_index'] = final_arts_finance_index['donation_index'].round(2)
final_arts_finance_index
```

## 예술재정지수

```python
final_arts_finance_index['arts_finance_index'] = final_arts_finance_index['budget_index'] + final_arts_finance_index['donation_index']
final_arts_finance_index['arts_finance_index'] = final_arts_finance_index['arts_finance_index']/2
final_arts_finance_index['arts_finance_index'] = final_arts_finance_index['arts_finance_index'].round(2)
final_arts_finance_index
```
