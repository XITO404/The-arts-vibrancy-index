## 문화예술활동가수

```python
# 계산을 위한 Table
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
# ((performanceTeam_2021 / performanceTeam_2021), (applicants_2023/applicants_2022))*100
cal_df['performanceTeam'] = art_manpower_index['performanceTeam_2021']/art_manpower_index['performanceTeam_2021']
cal_df['performanceTeam'] = cal_df['performanceTeam']*100
cal_df['applicants'] = art_manpower_index['applicants_2023']/art_manpower_index['applicants_2022']
cal_df['applicants'] = cal_df['applicants']*100
cal_df
```

```python
cal_df['performanceTeam_applicants'] = (cal_df['performanceTeam'] + cal_df['applicants'])/2
cal_df
```

```python
final_art_manpower_index['Arts_and_Culture_Activist'] = cal_df['performanceTeam_applicants']
final_art_manpower_index
```
