## 문화예술종사자수

```python
# 계산을 위한 Table
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
# (workers_2022/workers_2021)*100
cal_df['workers'] = arts_cp_index['workers_2022']/arts_cp_index['workers_2021']
cal_df['workers'] = cal_df['workers']*100
cal_df
```

```python
final['workers'] = cal_df['workers']
final
```

## 문화예술활동가수

```python
# 계산을 위한 Table
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
# ((groups_2021 / groups_2021)*100 + (applicants_2022 / applicants_2022)*100) / 2
# 2021, 2022 데이터를 사용해야 하나 해당 연도 데이터가 없으므로 100으로 나옴
cal_df['groups'] = arts_cp_index['groups_2021']/arts_cp_index['groups_2021']
cal_df['groups'] = cal_df['groups']*100
cal_df['applicants'] = arts_cp_index['applicants_2022']/arts_cp_index['applicants_2022']
cal_df['applicants'] = cal_df['applicants']*100
cal_df
```

```python
cal_df['participants'] = (cal_df['groups'] + cal_df['applicants'])/2
cal_df
```

```python
final['participants'] = cal_df['participants']
final
```

## 문화예술콘텐츠건수

```python
# 계산을 위한 Table
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
# (Content_2020/Content_2020)*100
cal_df['contents'] = arts_cp_index['contents_count_2020']/arts_cp_index['contents_count_2020']
cal_df['contents'] = cal_df['contents']*100
cal_df
```

```python
final['contents_count'] = cal_df['contents']
final
```

## 인력지수

```python
# 계산을 위한 Table
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
cal_df['manpower_index'] = final['workers'] + final['participants']
cal_df['manpower_index'] = cal_df['manpower_index']/2
cal_df
```

```python
final['manpower_index'] = cal_df['manpower_index']
final['manpower_index'] = final['manpower_index'].round(2)
final
```

## 콘텐츠지수

```python
cal_df['contents_index'] = final['contents_count']
cal_df['contents_index'] = cal_df['contents_index']
cal_df
```

```python
final['contents_index'] = cal_df['contents_index']
final
```

## 예술창작/생산지수

```python
final['arts_creation_production_index'] = final['manpower_index'] + final['contents_index']
final['arts_creation_production_index'] = final['arts_creation_production_index']/2
final['arts_creation_production_index'] = final['arts_creation_production_index'].round(2)
final
```
