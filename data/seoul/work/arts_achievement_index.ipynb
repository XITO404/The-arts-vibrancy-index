## 문화예술만족도 및 관람의향 (Arts and culture satisfaction and intent to attend)

```python
# 계산을 위한 Table
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
cal_df = pd.DataFrame(data)
cal_df
```

```python
# (Aacs_2022 / Aacs_2021, Avg(Aacitt_2022/Aacitt_2021))*100
cal_df['aacs'] = art_achievement_index['aacs_2022'] / art_achievement_index['aacs_2021']
cal_df['aacitt'] = (art_achievement_index['aacitt_2022'] / art_achievement_index['aacitt_2021']) / 9
cal_df['aacs_aacitt'] = (cal_df['aacs'] + cal_df['aacitt']) / 2 * 100
cal_df['aacs_aacitt'] = cal_df['aacs_aacitt'].round(2)
cal_df
```

```python
final_art_achievement_index['sat_intent'] = cal_df['aacs_aacitt']
final_art_achievement_index
```

## 문화예술교육만족도 및 수강의향

```python
# Avg((Aacs_2022 / Aacs_2021), Avg(Aacitt_2022/Aacitt_2021))*100
cal_df['swaace'] = art_achievement_index['swaace_2022']/art_achievement_index['swaace_2021']
cal_df['swaace'] = cal_df['swaace']*100
cal_df
```

```python
final_art_achievement_index['edu_intent'] = cal_df['swaace']
final_art_achievement_index['edu_intent'] = final_art_achievement_index['edu_intent'].round(2)
final_art_achievement_index
```

## 여가생활 만족도

```python
# (Lls_2022/Lls_2021)*100
cal_df['lls'] = art_achievement_index['lls_2022']/art_achievement_index['lls_2021']
cal_df['lls'] = cal_df['lls']*100
cal_df
```

```python
final_art_achievement_index['sat_leisure'] = cal_df['lls']
final_art_achievement_index['sat_leisure'] = final_art_achievement_index['sat_leisure'].round(2)
final_art_achievement_index
```

## 행복 및 삶의질 (Happiness and quality of life)

```python
# (happiness_2022/happiness_2021)
cal_df['happiness'] = art_achievement_index['happiness_2022']/art_achievement_index['happiness_2021']
cal_df['happiness'] = cal_df['happiness']*100
cal_df['qol'] = art_achievement_index['qof_2022']/art_achievement_index['qof_2021']
cal_df['qol'] = cal_df['qol']*100
cal_df['happiness_Qol'] = (cal_df['happiness'] + cal_df['qol'])/2
cal_df['happiness_Qol']
cal_df
```

```python
final_art_achievement_index['happy_life'] = cal_df['happiness_Qol']
final_art_achievement_index['happy_life'] = final_art_achievement_index['happy_life'].round(2)
final_art_achievement_index
```

## 공연시장매출액

```python
# (ms_2022/ms_2021)*100
cal_df['ms'] = art_achievement_index['ms_2022']/art_achievement_index['ms_2021']
cal_df['ms'] = cal_df['ms']*100
cal_df
```

```python
final_art_achievement_index['perf_revenue'] = cal_df['ms']
final_art_achievement_index['perf_revenue'] = final_art_achievement_index['perf_revenue'].round(2)
final_art_achievement_index
```

## 미술시장매출액

```python
# (Ams_2021/Ams_2021)*100 ----> 2022년 데이터가 존재하지 않기 때문에 지표 값이 1.00
cal_df['ams'] = art_achievement_index['ams_2021']/art_achievement_index['ams_2021']
cal_df['ams'] = cal_df['ams']*100
cal_df
```

```python
final_art_achievement_index['art_revenue'] = cal_df['ams']
final_art_achievement_index['art_revenue'] = final_art_achievement_index['art_revenue'].round(2)
final_art_achievement_index
```

## 문화예술 국제교류 건수

```python
# (Noce_2019/Noce_2019)*100 ----> 2019년 데이터만 존재하지 않기 때문에 지표 값이 1.00
cal_df['noce'] = art_achievement_index['noce_2019']/art_achievement_index['noce_2019']
cal_df['noce'] = cal_df['noce']*100
cal_df
```

```python
final_art_achievement_index['global_exchange'] = cal_df['noce']
final_art_achievement_index
```

## 향유결과지수

```python
cal_df['enjoyment_outcome_index'] = final_art_achievement_index['sat_intent'] + final_art_achievement_index['edu_intent'] + final_art_achievement_index['sat_leisure'] + final_art_achievement_index['happy_life']
cal_df['enjoyment_outcome_index'] = cal_df['enjoyment_outcome_index'] / 4
cal_df
```

```python
final_art_achievement_index['enjoyment_outcome_index'] = cal_df['enjoyment_outcome_index']
final_art_achievement_index['enjoyment_outcome_index'] = final_art_achievement_index['enjoyment_outcome_index'].round(2)
final_art_achievement_index
```

## 시장성과지수

```python
cal_df['market_performance_index'] = final_art_achievement_index['perf_revenue'] + final_art_achievement_index['art_revenue'] + final_art_achievement_index['global_exchange']
cal_df['market_performance_index'] = cal_df['market_performance_index']/3
cal_df
```

```python
final_art_achievement_index['market_performance_index'] = cal_df['market_performance_index']
final_art_achievement_index['market_performance_index'] = final_art_achievement_index['market_performance_index'].round(2)
final_art_achievement_index
```

```python
final_art_achievement_index['arts_achievement_index'] = final_art_achievement_index['enjoyment_outcome_index'] + final_art_achievement_index['market_performance_index']
final_art_achievement_index['arts_achievement_index'] = final_art_achievement_index['arts_achievement_index']/2
final_art_achievement_index['arts_achievement_index'] = final_art_achievement_index['arts_achievement_index'].round(2)
final_art_achievement_index
```
