```python
import pandas as pd
```

```python
art_facilities_index = pd.read_csv('drive/MyDrive/ArtIndex/busan_facilities_data.csv')
art_facilities_index
```

```python
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}

final_art_facilities_index = pd.DataFrame(data)
final_art_facilities_index
```

```python
data = {
    '구': ['강서구', '금정구', '기장군', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구', '영도구', '중구', '해운대구'],
}

cal_df = pd.DataFrame(data)
cal_df
```

```python
cal_df['hall_num'] = art_facilities_index['hall_num']
cal_df['mil_fac_num'] = art_facilities_index['mil_fac_num']
cal_df['hall_num'] = art_facilities_index['hall_num']
cal_df['hall_rate'] = art_facilities_index['hall_rate']
cal_df['hall_people'] = art_facilities_index['hall_people']
cal_df['art_people'] = art_facilities_index['art_people']
```

```python
# 시설지수 계산
cal_df['hall_num_rate'] = (cal_df['hall_num'] + cal_df['hall_rate'])/2
```

```python
final_art_facilities_index['facilities_index'] = (cal_df['hall_num'] + cal_df['mil_fac_num']+ cal_df['hall_num_rate'])/3
final_art_facilities_index
```

```python
# 시설인력지수 계산
final_art_facilities_index['facilities_people_index'] = (cal_df['hall_people'] + cal_df['art_people'])/2
final_art_facilities_index
```

```python
# 예술시설지수
final_art_facilities_index['art_facilities_index'] = final_art_facilities_index['facilities_index'] + final_art_facilities_index['facilities_people_index']
final_art_facilities_index['art_facilities_index'] = final_art_facilities_index['art_facilities_index']/2
final_art_facilities_index
```

```python
final_art_facilities_index.to_csv('drive/MyDrive/ArtIndex/busan_art_facilities_index.csv', encoding='utf-8-sig', index=False)
```
