```python
# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
Seoul_arts_index = pd.DataFrame(data)
Seoul_arts_index
```

```python
arts_creation_production_index = pd.read_csv('/content/arts_creation_production_index.csv')
arts_finance_index = pd.read_csv('/content/arts_finance_index.csv')
arts_facilities_index = pd.read_csv('/content/arts_facilities_index.csv')
arts_enjoyment_index = pd.read_csv('/content/arts_enjoyment_index.csv')
arts_achievement_index = pd.read_csv('/content/arts_achievement_index.csv')
```

## 예술창작/생산지수

```python
arts_creation_production_index
```

```python
Seoul_arts_index['문화예술종사자 수'] = arts_creation_production_index['workers'].round(2)
Seoul_arts_index['예술 활동가 수'] = arts_creation_production_index['participants'].round(2)
Seoul_arts_index['문화예술 콘텐츠 건수'] = arts_creation_production_index['contents_count'].round(2)
Seoul_arts_index['문화예술 콘텐츠 횟수'] = 0
Seoul_arts_index['예술창작/생산지수'] = arts_creation_production_index['arts_creation_production_index'].round(2)
Seoul_arts_index
```

## 예술재정지수

```python
arts_finance_index
```

```python
Seoul_arts_index['문화예술 부서 예산'] = arts_finance_index['cul_budget'].round(2)
Seoul_arts_index['문화예술 예산'] = arts_finance_index['cna_budget'].round(2)
Seoul_arts_index['문화예술 부서 예산 대비 문화예술 예산'] = arts_finance_index['cna_budget_2'].round(2)
Seoul_arts_index['전체 예산 대비 문화예술 부서 예산'] = arts_finance_index['cul_government'].round(2)
Seoul_arts_index['기업의 문화예술 지원액'] = 0
Seoul_arts_index['기업의 문화예술 기부액'] = arts_finance_index['donation'].round(2)
Seoul_arts_index['예술재정지수'] = arts_finance_index['arts_finance_index'].round(2)
Seoul_arts_index
```

## 예술시설지수

```python
arts_facilities_index
```

```python
Seoul_arts_index['문화시설 수'] = arts_facilities_index['cul_fac_num'].round(2)
Seoul_arts_index['인구 100만 명당 문화시설 수'] = arts_facilities_index['mil_fac_num'].round(2)
Seoul_arts_index['공연장 수 및 가동률'] = arts_facilities_index['hall_num_rate'].round(2)
Seoul_arts_index['공연시설 인력 수'] = arts_facilities_index['hall_people'].round(2)
Seoul_arts_index['미술시설 인력 수'] = arts_facilities_index['art_people'].round(2)
Seoul_arts_index['예술시설지수'] = arts_facilities_index['arts_facilities_index'].round(2)
Seoul_arts_index
```

## 예술향유지수

```python
arts_enjoyment_index
```

```python
Seoul_arts_index['문화예술 관람률'] = arts_enjoyment_index['audience_rate'].round(2)
Seoul_arts_index['문화예술 관람횟수'] = arts_enjoyment_index['visit_num'].round(2)
Seoul_arts_index['가구당 문화서비스지출'] = arts_enjoyment_index['expenditure'].round(2)
Seoul_arts_index['독서 및 도서관 이용률'] = arts_enjoyment_index['reading'].round(2)
Seoul_arts_index['문화예술분야 창작발표 경험률'] = 0
Seoul_arts_index['문화예술분야 자원봉사 경험률'] = arts_enjoyment_index['volunteer'].round(2)
Seoul_arts_index['문화예술분야 동호회 참여율'] = arts_enjoyment_index['club'].round(2)
Seoul_arts_index['문화예술교육 경험률'] = arts_enjoyment_index['education'].round(2)
Seoul_arts_index['예술향유지수'] = arts_enjoyment_index['arts_enjoyment_index'].round(2)
Seoul_arts_index
```

## 예술성과지수

```python
arts_achievement_index
```

```python
Seoul_arts_index['문화예술만족도 및 관람의향'] = arts_achievement_index['sat_intent'].round(2)
Seoul_arts_index['문화프로그램 만족도 및 수강의향'] = arts_achievement_index['edu_intent'].round(2)
Seoul_arts_index['여가생활 만족도'] = arts_achievement_index['sat_leisure'].round(2)
Seoul_arts_index['행복수준 및 삶의 질'] = arts_achievement_index['happy_life'].round(2)
Seoul_arts_index['공연시장 매출액'] = arts_achievement_index['perf_revenue'].round(2)
Seoul_arts_index['미술시장 매출액'] = arts_achievement_index['art_revenue'].round(2)
Seoul_arts_index['문화예술 국제교류 건수'] = arts_achievement_index['global_exchange'].round(2)
Seoul_arts_index['예술성과지수'] = arts_achievement_index['arts_achievement_index'].round(2)
Seoul_arts_index
```

## Arts Index

```python
Seoul_arts_index['서울예술활력지수'] = arts_creation_production_index['arts_creation_production_index'] + arts_finance_index['arts_finance_index'] + arts_facilities_index['arts_facilities_index'] + arts_enjoyment_index['arts_enjoyment_index'] + arts_achievement_index['arts_achievement_index']
Seoul_arts_index['서울예술활력지수'] = Seoul_arts_index['서울예술활력지수']/5
Seoul_arts_index['서울예술활력지수'] = Seoul_arts_index['서울예술활력지수'].round(2)
Seoul_arts_index
```

```python
import matplotlib.pyplot as plt

# 폰트 설정
plt.rcParams['font.family'] = 'NanumBarunGothic'

# 새로운 figure 생성 및 크기 설정
plt.figure(figsize=(10, 6))  # 가로크기 10, 세로크기 6

Seoul_arts_index['서울예술활력지수'].plot(kind='bar')
plt.title('서울예술활력지수')

# x축에 구별 이름 추가
districts = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']
plt.xticks(range(len(districts)), districts, rotation=45, ha='right')

plt.show()
```
