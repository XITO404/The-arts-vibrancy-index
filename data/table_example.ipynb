```python
import pandas as pd

# 서울시 구별 데이터
data = {
    '구': ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'],
}
# 데이터 프레임 생성
seoul_df = pd.DataFrame(data)
seoul_df
```

```python
# 지표 파일
seoul_happiness = pd.read_excel('seoul_happiness.xlsx')
```

```python
seoul_happiness = seoul_happiness.sort_values(by='구', ascending=True)
seoul_happiness = seoul_happiness.reset_index(drop=True)  # 인덱스 다시 지정
seoul_happiness
```

```python
seoul_df['happiness_2021'] = seoul_happiness[2021]
seoul_df['happiness_2022'] = seoul_happiness[2022]
seoul_df
```

```python
seoul_df.to_csv('seoul_data.csv', index=False)
```
