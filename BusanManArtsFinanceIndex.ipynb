```python
# 문화부 대비 문화예술 비율 구하는
import pandas as pd

# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장구', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구','중구', '해운대구'],
}
# 데이터 프레임 생성
busan_df = pd.DataFrame(data)
busan_df

busan_CultureBudget = pd.read_excel('BusanCultureBudget.xlsx', engine='openpyxl')  #문화부 예산
busan_CulArtBudget = pd.read_excel('BusanCulArtBudget.xlsx', engine='openpyxl')#문화예술 예산

busan_CultureBudget = busan_CultureBudget.sort_values(by='구', ascending=True)

busan_CultureBudget = busan_CultureBudget.reset_index(drop=True)  # 인덱스 다시 지정
busan_CulArtBudget = busan_CulArtBudget.reset_index(drop=True)

busan_df['2021'] = busan_CulArtBudget[2021]/busan_CultureBudget[2021]*100
busan_df['2022'] = busan_CulArtBudget[2022]/busan_CultureBudget[2022]*100


busan_df.to_csv('busanCul_Art.csv',encoding='utf-8-sig', index=False) #문화부예산 대비 문화예술 예산
busan_df
```

```python
#BusanFinance테이블 만들기

# 부산시 구별 데이터
data = {
    '구': ['강서구', '금정구', '기장구', '남구', '동구', '동래구', '부산진구', '북구', '사상구', '사하구', '서구', '수영구', '연제구','중구', '해운대구'],
}
# 데이터 프레임 생성
busan_df = pd.DataFrame(data)
busan_df

busan_CultureBudget = pd.read_excel('BusanCultureBudget.xlsx', engine='openpyxl')  #문화부 예산
busan_CulArtBudget = pd.read_excel('BusanCulArtBudget.xlsx', engine='openpyxl')#문화예술 예산
busan_BusanCul_Goverment = pd.read_excel('BusanCul_Goverment.xlsx', engine='openpyxl')#정부대비 문화부예
busan_Cul_Art = pd.read_excel('busanCul_Art.xlsx', engine='openpyxl')#문화부대비 문화예술 예산
busan_BusanDonation= pd.read_excel('BusanDonation.xlsx', engine='openpyxl')#부산문화기부금

busan_CultureBudget = busan_CultureBudget.sort_values(by='구', ascending=True)

busan_CultureBudget = busan_CultureBudget.reset_index(drop=True)  # 인덱스 다시 지정
busan_CulArtBudget = busan_CulArtBudget.reset_index(drop=True)
busan_BusanCul_Goverment = busan_BusanCul_Goverment.reset_index(drop=True)
busan_Cul_Art = busan_Cul_Art.reset_index(drop=True)
busan_BusanDonation =busan_BusanDonation.reset_index(drop=True)

busan_df['busan_CultureBudget2021'] =busan_CultureBudget[2021]
busan_df['busan_CultureBudget2022'] =busan_CultureBudget[2022]

busan_df['busan_CulArtBudget2021'] =busan_CulArtBudget[2021]
busan_df['busan_CulArtBudget2022'] =busan_CulArtBudget[2022]

busan_df['busan_Cul_Goverment2021'] =busan_BusanCul_Goverment[2021]
busan_df['busan_Cul_Goverment2022'] =busan_BusanCul_Goverment[2022]

busan_df['busan_Cul_Art2021'] =busan_Cul_Art[2021]
busan_df['busan_Cul_Art2022'] =busan_Cul_Art[2022]

busan_df['busan_BusanDonation2021'] =busan_BusanDonation[2021]
busan_df['busan_BusanDonation2022'] =busan_BusanDonation[2022]

busan_df.to_csv('arts_finance_calculate.csv',encoding='utf-8-sig', index=False)
busan_df
```

```python
#BusanFinanceIndex


busan_df['busan_CultureBudget2021'] =busan_CultureBudget[2021]
busan_df['busan_CultureBudget2022'] =busan_CultureBudget[2022]

busan_df['busan_CulArtBudget2021'] =busan_CulArtBudget[2021]
busan_df['busan_CulArtBudget2022'] =busan_CulArtBudget[2022]

busan_df['busan_Cul_Goverment2021'] =busan_BusanCul_Goverment[2021]
busan_df['busan_Cul_Goverment2022'] =busan_BusanCul_Goverment[2022]

busan_df['busan_Cul_Art2021'] =busan_Cul_Art[2021]
busan_df['busan_Cul_Art2022'] =busan_Cul_Art[2022]

busan_df['busan_BusanDonation2021'] =busan_BusanDonation[2021]
busan_df['busan_BusanDonation2022'] =busan_BusanDonation[2022]


busan_df.to_csv('BusanArtFinanceIndex.csv',encoding='utf-8-sig', index=False)
busan_df
```

```python
# 데이터 프레임 생성
busan_df_index = pd.DataFrame(data)
busan_df_index

#예산관련지수들 정리
busan_df_index['CultureBudget']=busan_CultureBudget[2022]/busan_CultureBudget[2021]*100
busan_df_index['Cul_Goverment']=busan_BusanCul_Goverment[2022]/busan_BusanCul_Goverment[2021]*100
busan_df_index['CulArtBudget']=busan_CulArtBudget[2022]/busan_CulArtBudget[2021]*100
busan_df_index['Cul_Art']=busan_Cul_Art[2022]/busan_Cul_Art[2021]*100

#기부지수
busan_df_index['donation_index']=busan_BusanDonation[2022]/busan_BusanDonation[2021]*100

#예산지수
busan_df_index['budget_index'] = (
    busan_df_index['CultureBudget']+busan_df_index['Cul_Goverment']+busan_df_index['CulArtBudget']+busan_df_index['Cul_Art']
)/4


#재정지수(예산+기부)
busan_df_index['finanace_index']=(busan_df_index['donation_index']+busan_df_index['budget_index'])/2
busan_df_index

busan_df_index.to_csv('Art_FinanceIndex.csv',encoding='utf-8-sig', index=False)
```
